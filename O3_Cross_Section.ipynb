```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
o3 = np.transpose(np.genfromtxt('./spectra/O3_cross_section.dat'))
```

```python
plt.figure(figsize=(12,8))
for i in range(1,len(o3)):
    plt.plot(o3[0],o3[i])

plt.yscale('log')
```

```python
o3_cut = o3[:,np.logical_and(o3[0]>=380, o3[0]<=780)]
```

```python
plt.plot(o3_cut[0],o3_cut[1])
plt.yscale('log')
```

```python
np.savetxt('./spectra/O3_cross_section_cut.dat',np.transpose(o3_cut))
```
